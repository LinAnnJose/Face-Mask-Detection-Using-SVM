```python
flat_data_arr=[] #collect input samples
target_arr=[] #Collect output samples
categories=['WithMask','WithoutMask']
dirpath='/content/drive/MyDrive/MaskFiles'
```

```python
for i in categories:
  path=os.path.join(dirpath,i)
  for img in os.listdir(path):
    img_arr=imread(os.path.join(path,img))
    img_resize=resize(img_arr,(150,150,3))
    flat_data_arr.append(img_resize.flatten())
    target_arr.append(categories.index(i))
  print("Loaded completely",i)

```

```python
#Create arrays of input and outputs
flat_data=np.array(flat_data_arr)
target_data=np.array(target_arr)
#Create dataframe from input dsamples
df=pd.DataFrame(flat_data)
df.shape
```

```python
#Add outputsample to dataframe
df['Target']=target_data
df.shape
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
model=SVC()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy score is",accuracy_score(y_test,y_pred))
```
